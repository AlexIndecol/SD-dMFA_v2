Load indicator outputs from a run folder and visualise them.

Run a dry-run first:

```bash
crm-model run --scenario baseline --dry-run
```

````python
from pathlib import Path
import pandas as pd
import yaml
import matplotlib.pyplot as plt

def find_repo_root(start: Path) -> Path:
    for p in [start, *start.parents]:
        if (p/"configs"/"time.yml").exists() and (p/"data"/"exogenous").exists():
            return p
    raise FileNotFoundError("Could not locate repo root containing configs/ and data/exogenous/")

ROOT = find_repo_root(Path(".").resolve())
CFG = ROOT/"configs"
DATA = ROOT/"data"/"exogenous"

def load_yaml(path):
    with open(path, "r", encoding="utf-8") as f:
        return yaml.safe_load(f)


OUT = ROOT/'outputs'
# Pick the latest run folder
runs = sorted(OUT.glob('baseline/*'))
if not runs:
    raise SystemExit('No runs found. Run: crm-model run --scenario baseline --dry-run')
run_dir = runs[-1]
print('Using run:', run_dir)
ind_dir = run_dir/'indicators'
ind_files = sorted(ind_dir.glob('*.csv'))
print('Indicators:', [p.stem for p in ind_files])
````
